```python
from Operations_2 import *
import matplotlib.pyplot as plt
import pandas as pd

# Declare Class
readSheet=ReadSheet()

# Declare function
sheetHList=readSheet.Authorization_Hotel()

print("------------- Hotel Price Monitoring ------------")    
print("")

count=1
for n in sheetHList:
    dfIn=readSheet.LoadSheet(n)
    #display(dfIn)

    priceList=[]
    for n in dfIn["Price"]:
        #print(n," :: ",type(n),' :: ',float(n.replace(',','')))
        priceList.append(float(n.replace(',','')))
    dateList=[]
    for n in dfIn["Date"]:
        #print(n," :: ",type(n),' :: ')
        #dateList.append(n.strftime("%Y-%m-%d"))
        dateList.append(n)

    if(count==1):
        titleString=" Hotel Labaris - Feb 15-17 2021"
    else:
        titleString=" Oakwood Sriracha - Feb 15-17 2021"
        
    

    dfIn_1=dfIn[['Date','Price']].set_index('Date')
    dfIn_1['Price'] = dfIn_1['Price'].str.replace(',', '').astype(float)

    fig, ax = plt.subplots(figsize=(7, 4))

    ax.plot(dfIn_1.index.values,
            dfIn_1['Price'],
            color='purple')

    ax.set(xlabel="Date",
           ylabel="Price",
           title=titleString)

    plt.setp(ax.get_xticklabels(), rotation=45)

    plt.show()
    count+=1
    
```
